```python
import pandas as pd
```

```python
path_to_csv = "DER_report.csv"

df = pd.read_csv(path_to_csv)
df.head()
```

```python
# get all the values in the name column
output = {"base":[], "id":[]}
names = df['name'].values
for name in names:
    try:
        base = name.split("_")[0]
        if base == "GERO" or base == "Slachevsky":
            id = name.split("_")[1] + "_" + name.split("_")[2]
        else:
            id = name.split("_")[1]
        print(base, id)
        # save to output
        output["base"].append(base)
        output["id"].append(id)
        
    except:
        continue
```

```python
output = pd.DataFrame(output)

output.to_csv("output.csv", index=False)
```
