```python
import os
import pandas as pd
from numpy.random import default_rng
rng = default_rng()

# get the names of the audios that have already been diarized
path_to_csv = "DER_report.csv"

df = pd.read_csv(path_to_csv)
names = df['name'].values
```

```python
base_completa = pd.read_csv("FONDECYT+Slachevsky+CETRAM+GERO.csv")
```

```python
for destination_folder in os.listdir("todo"):
    i = 0
    amount = destination_folder.split("_")[-1]
    base_name = destination_folder.split("_")[0]
    desired_group = destination_folder.split("_")[1]
    path_to_folder = f"datasets/{base_name}_converted"

    print("Looking for ", amount, " audios from group ", desired_group, " in ", path_to_folder)

    while i <= int(amount):
        # pick a random file in folder
        audio = rng.choice(os.listdir(path_to_folder), size=1, replace=False)
        audio = audio[0]
        if audio.split(".")[0] + ".txt" in names[2:]:
            i -= 1
        else:
            try:
                #check if audio belongs to group
                base = audio.split("_")[0]
                if base == "GERO" or base == "Slachevsky":
                    id = audio.split("_")[1] + "_" + audio.split("_")[2]
                else:
                    id = audio.split("_")[1]
                # with base and id we can get the group from the base_completa
                
                group = base_completa[base_completa['id'] == id]['group'].values[0]
                if group == desired_group:
                    print("Moving ", audio, "to the right folder")
                    # copy the audio to the right folder
                    out =  os.path.join("todo", destination_folder)
                    os.system(f"cp {path_to_folder}/{audio} {out}")
                else:
                    i -= 1
            except:
                i -= 1
        i += 1
```
