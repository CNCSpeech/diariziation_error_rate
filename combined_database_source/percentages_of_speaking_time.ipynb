# speaker percents

the idea is to obtain a percentage of speaking time from each speaker in each audio, and then average them out.

```python
# the idea is to obtain a percentage of speaking time from each speaker in each audio, and then average them out.
# audios are in diffent csvs in a folder

import pandas as pd
import os
import glob
```

```python
# path to csvs

diarization_paths = ["datasets/Slachevsky_converted/diarization",
                     "datasets/GERO_renamed/diarization",
                     "datasets/FONDECYT_converted/diarization",
                     "datasets/CETRAM_converted/diarization"]

# get all csvs in the folder
print("Average percentage of speaking time of main speaker \n ")
for diarization_path in diarization_paths:
    all_csvs = glob.glob(diarization_path + "/*.csv")

    # get the number of speakers in each audio
    speakers = {}
    for csv in all_csvs:
        df = pd.read_csv(csv)
        # get the number of speakers and their total speaking time
        num_speakers = df['speaker'].nunique()
        # sum the speaking time of each speaker
        speaking_time = df.groupby('speaker')['dur [s]'].sum()
        # get the percentage of speaking time of each speaker
        speaking_time_percentage = speaking_time / speaking_time.sum()
        # add the percentage of speaking time of each speaker to the dictionary
        speakers[csv] = speaking_time_percentage

    # get the average percentage of the speaking time of the maximum speaker in each audio
    speaking_time_percentage = {}
    for csv in speakers:
        speaking_time_percentage[csv] = speakers[csv].max()

    # get the average percentage of the speaking time of the maximum speaker in all audios
    speaking_time_percentage = pd.Series(speaking_time_percentage)
    average_speaking_time_percentage = speaking_time_percentage.mean()
    # stdev, median
    std = speaking_time_percentage.std()
    # print the results
    print(diarization_path.split("/")[1], "mean: ", average_speaking_time_percentage, "std: ", std,"\n")
```
